```python
import pandas as pd
```

```python
TITANIC_FILE_PATH = 'titanic.csv'
```

```python
df = pd.read_csv(TITANIC_FILE_PATH )
```

```python
df.head()
```

# 범주형 데이터 Sex, Embarked를 수치형 데이터로

```python
titanic_sex_embarked = df[['Sex', 'Embarked']]
```

```python
titanic_sex_embarked.head()
```

```python
one_hot_encoded_df = pd.get_dummies(titanic_sex_embarked)
```

```python
one_hot_encoded_df.head()
```

# 더 복잡한 방법

```python
one_hot_encoded_df = pd.get_dummies(data=df, columns=['Sex', 'Embarked'])
one_hot_encoded_df.head()
```
